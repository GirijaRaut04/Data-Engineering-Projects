```python
import json
import pandas as pd

# Load the raw JSON file
with open("335982.json", "r") as f:
    data = json.load(f)
```

```python
# Initialize a list to hold flattened delivery-level records
balls_data = []

# Loop through each inning
for inning in data.get("innings", []):
    team = inning.get("team")  # Get the team name batting in the inning
    
    # Loop through overs in the inning
    for over_entry in inning.get("overs", []):
        over_number = over_entry.get("over")  # e.g., 0, 1, 2...

        # Loop through each ball (delivery) in the over
        for ball in over_entry.get("deliveries", []):
            # Extract key delivery information with defaults
            row = {
                "team": team,
                "over": over_number,
                "batter": ball.get("batter"),
                "bowler": ball.get("bowler"),
                "non_striker": ball.get("non_striker"),
                "runs_batter": ball.get("runs", {}).get("batter", 0),
                "runs_extras": ball.get("runs", {}).get("extras", 0),
                "runs_total": ball.get("runs", {}).get("total", 0),
                "extras": ball.get("extras", {}),          # e.g., {'wides': 1}
                "wickets": ball.get("wickets", [])          # could be empty or contain dismissal info
            }
            balls_data.append(row)  # Append to list
```

```python

# Convert the list of deliveries to a DataFrame
df = pd.DataFrame(balls_data)

# (Optional) You could apply further filtering here, e.g.:
# df = df[df['runs_total'] > 0]  # keep only balls with runs

# Convert the cleaned DataFrame back to JSON format
inning_json = df.to_dict(orient="records")

# Save to a new JSON file
with open("inning_335982.json", "w") as f:
    json.dump(inning_json, f, indent=2)
```
